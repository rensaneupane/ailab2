```python
import pandas as pd
import seaborn  as sns
import matplotlib.pyplot as plt
```

```python
def load_dataset(file_path):
  return pd.read_csv(file_path)


df = load_dataset("/content/aREN).zip")
df.head(20)
```

```python
sns.countplot(data=df, x='experience_level')
df['experience_level'].value_counts()
```

```python
 # pie chart

df['experience_level'].value_counts().plot(kind='pie',autopct='%0.3f')
```

```python
# Numberical Value

# Histogram treat as categorical data distribution of data

plt.hist(df['experience_level'])
```

```python
# Distplot
# probability density function --> The line is kde
# skweness
sns.distplot(df['salary_in_usd'])
```

```python
# Boxplot
# 5 number summary
# median , percentile, min=Q1-1.5*IQR, max=Q3+1.5*IQR, IQR = (Q1-Q3), outlier

sns.boxplot(df['salary_in_usd'])

```

```python
# Bar plot (Numerical - Categorical)
  # average age in pclass
sns.barplot(x='salary',y='salary_currency',hue='job_title',data=df.head(20))
```

```python
sns.heatmap(pd.crosstab(df.head(20)['employee_residence'],df.head(20)['job_title']))
```

```python
sns.pairplot(df.head(25),hue='job_title')
```
